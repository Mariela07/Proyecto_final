```python
df_clean = df[(df["Debt_to_Equity_Ratio"]<=30) & (df["Interest_Coverage_Ratio"]<=1400) & (df["Current_Ratio"]<=200)]
df_clean.shape
```

```python
df_clean["Total_Revenue"].fillna(df_clean["Total_Revenue"].mean(), inplace=True)
df_clean["Short_Term_Debt"].fillna(df_clean["Short_Term_Debt"].mean(), inplace=True)
df_clean["Financial_Expenses"].fillna(df_clean["Financial_Expenses"].mean(), inplace=True)
```
